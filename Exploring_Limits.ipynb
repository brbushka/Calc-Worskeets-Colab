The concept of a <b>limit</b> is one of the most essential tools in Caclulus. Not only are limits important in their own right (give us a convenient way to describe how functions behave), but they also are the foundation of the other fundamental calculus concepts, such as differentiation and integration.    
<br>

<div style="background-color:lightgray">
<b>Definition of a Limit (intuitive):</b>

Suppose the function $f$ is defined for all $x$ near the number $a$, except possibly at $a$ itself. If we can make the values of $f(x)$ arbitrarily close to $L$ (as close to $L$ as we like) by restricting $x$ to be sufficiently close to $a$ (but not equal to $a$), we write
  
$$\displaystyle{\lim_{x\rightarrow a}f(x)}=L$$
   
and say "the limit of $f(x)$, as $x$ approaches $a$ equals $L$".
</div>

<font size="4"><span style='color:green'> <b>Example 1.</b></span></font> Consider the function $$f(x)=\dfrac{\sqrt{x}-1}{x-1}.$$
(a) Create a table of values of $f$ corresponding to values of $x$ near $1$.<br>
(b) Speculate whether $\displaystyle{\lim_{x\rightarrow 1}f(x)}$ appears to exist, and if it does, make a conjecture about its value. <br>
(c) Plot the graph of $f$.<br>
(d) Does the graph appear to be consistent with your findings in (b).<br>

<span style='color:green'> <b>Note 1:</b></span> The domain of $f$ is $\{x|x\neq 0\}$. While $f$ is <i>not</i> defined for $x=1$, we want to see how the function behaves when $x$ is in the neighborhood $1$.  

<br>
<span style='color:orange'> <b>Solution:</b></span> In order to accomplish the above tasks, we'll use Python to create the numerical table and plot.

```python
#Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
from IPython.display import Latex, display
```

```python
#Define the function in Python
def f(x):
  return (np.sqrt(x)-1)/(x-1)
```

(a) Our first task is to create a table of values for $f$ corresponding to inputs of $x$ near $1$.

```python
#Make a list of inputs near 1. As we go from left to right, we are getting closer and closer to 1.
inputsL = [0.9, 0.99, 0.999, 0.9999, 0.99999, 0.999999,0.9999999]

#Make a list of the corresponding outputs
outputsL = [f(x) for x in inputsL]

print(tabulate([inputsL, outputsL],floatfmt=".9f",tablefmt='fancy_grid'))
#todo: FIX THE HEADERS
```

From the table, it appears that as $x$ approaches $1$ from the left, the function outputs are getting closer and closer to $0.5$. Symbollically, we can write this as:

<br>

<center> $\displaystyle{\lim_{x \rightarrow 1^-}f(x)}=0.5$</center>
or
<center> as $ x\rightarrow 1^-$, $f(x) \rightarrow 0.5$.</center>

Next, we'll repeat the above process, but our inputs will be aproaching 1 from the right this time.

```python
#Make a list of inputs as x is approaching 1 from the right. Read the table from right to left.
inputsR = [1.0000001, 1.000001, 1.00001, 1.0001, 1.001, 1.01, 1.1]
outputs = [f(x) for x in inputsR]
print('***Read the table from right to left***')
#The "0.8" specifies the number of decimal places to be displayed for the floating-point numbers in the table
print(tabulate([inputsR, outputs],floatfmt=".8f"))
```

Again, from this table, it appears that as $x$ approaches $1$ from the right, the function outputs are getting closer and closer to $0.5$. So we say:
<br>

<br>
<center> $\displaystyle{\lim_{x \rightarrow 1^+}f(x)}=0.5$</center>
or
<center> as $ x\rightarrow 1^+$, $f(x) \rightarrow 0.5$.</center>

The numerical evidence suggests that the function outputs appear to get closer and closer to $0.5$, as $x$ aproaches 1 from either direction. Therefore, we can make a conjecture that $\displaystyle{\lim_{x \rightarrow 1}f(x)}=0.5$.

(b) Our next task is to plot the graph of $f$. Once again, we will do that with the help of Python.

```python
def f(x):
    return (np.sqrt(x)-1)/(x-1)

#in this case linspace takes 50 values from 0, up to 5 (but not including 5)
x = np.linspace(0, 5, 50)
y = f(x)

plt.plot(x, y,'o')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plotting some points on the graph of y=f(x)")
plt.show()
```

```python
#import numpy as np
#import matplotlib.pyplot as plt

def f(x):
    return (np.sqrt(x)-1)/(x-1)

x = np.linspace(0, 5, 100)
y = f(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of f(x) = (sqrt(x)-1)/(x-1)")
plt.show()
```

<span style='color:green'> <b>Note 3:</b></span> Notice that from the graph it appears that the function output at 1 is 0.5, but we know that $x=1$ is not in the domain of $f$. If we try to exaluate $f(1)$ in Python, we'll get an error (See below).   

```python
# Evaluate the function at 1
# nan in Python stands for Not a Number

print(f(1))
```

```python
#Colorcoding the parts of the graph from the left and right of 1

#def f(x):
#    return (np.sqrt(x)-1)/(x-1)

x_L = np.linspace(0, 0.99999, 100)
y_L = f(x_L)

x_R = np.linspace(1.00001, 5, 100)
y_R = f(x_R)


plt.plot(x_L, y_L)
plt.plot(x_R, y_R)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of f(x) = (sqrt(x)-1)/(x-1)")
plt.show()
```

(d) The table and the graph appear to be consistant. Since the function is not defined at 1 and $\displaystyle{\lim_{x\rightarrow 1^-}f(x)}=\displaystyle{\lim_{x\rightarrow 1^+}f(x)}=0.5$, we say that the graph of $y=f(x)$ has a *hole (removable discontinuity)* at $x=1$.

<font size="4"><span style='color:green'> <b>Example 2.</b></span></font> Investigate $\displaystyle{\lim_{x\rightarrow 0}\cos\left(\dfrac{1}{x}\right)}$ both numerically and graphically. Make a conjecture about its value, or state that the limit does not exist.  

```python
def g(x):
    return np.cos(1/x)

#Make a list of inputs as x is approaching 1 from the right. Read the table from right to left.
inputs_gL = [-0.1, -0.01, -0.001,-0.0001, -0.00001, -0.000001, -0.0000001,-0.00000001]
outputs_gL = [g(x) for x in inputs_gL]
print('***Read the table from left to right***')
print(tabulate([inputs_gL, outputs_gL],floatfmt=".9f"))

inputs_gR = [0.00000001,0.0000001, 0.000001, 0.00001, 0.0001, 0.001, 0.01, .1]
outputs_gR = [g(x) for x in inputs_gR]
print('***Read the table from right to left***')
print(tabulate([inputs_gR, outputs_gR],floatfmt=".9f"))
```

Notice that as $x$ approaches $0$ from either direction, the outputs do not seem to stabilize to a particular value. From this we can speculate that the given limit does not exist.

```python
#Colorcodeing the parts of the graph from the left and right of 1

#def g(x):
#    return (np.cos(1/x))

x_gL = np.linspace(0.000001, 1, 1000)
y_gL = g(x_gL)

x_gR = np.linspace(-1,-0.000001, 1000)
y_gR = g(x_gR)


plt.plot(x_gL, y_gL)
plt.plot(x_gR, y_gR)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of g(x) = cos(1/x)")
plt.show()
```

Zooming in on the above graph a little closer to $x=0$.

```python
#Colorcodeing the parts of the graph from the left and right of 1

#def g(x):
#    return (np.cos(1/x))

x_gL = np.linspace(0.000001, 0.01, 1000)
y_gL = g(x_gL)

x_gR = np.linspace(-0.01,-0.000001, 1000)
y_gR = g(x_gR)


plt.plot(x_gL, y_gL)
plt.plot(x_gR, y_gR)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of g(x) = cos(1/x)")
plt.show()
```

From the graph it appears that the graph is contained between $y=-1$ and $y=1$, but the function outputs keep oscillating and don't seem to settle to any particular value. To make a bit more sense of this, think of what happens as $x\rightarrow 0$. As $x\rightarrow 0$, $\dfrac{1}{x} \rightarrow \infty$; and thinking about the "ends" of the cosine function (as its input grows indefinitly), it makes sense that the above limit keeps bouncing between $-1$ and $1$.
